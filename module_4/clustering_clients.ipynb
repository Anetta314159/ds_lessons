### Описание признаков
##### Демографические данные

    CustomerID: Уникальный идентификатор клиента.
    BirthYear: Год рождения клиента.
    EducationLevel: Уровень образования клиента.
    MaritalStatus: Семейное положение клиента.
    AnnualIncome: Ежегодный доход домохозяйства клиента.
    NumChildren: Количество детей в семье клиента.
    NumTeenagers: Количество подростков в семье клиента.
    EnrollmentDate: Дата регистрации клиента в компании.

##### Поведение

    DaysSinceLastPurchase: Количество дней с момента последней покупки клиента.
    FiledComplaint: Указано, подавал ли клиент жалобы за последние 2 года (1 — да, 0 — нет).

##### Траты на продукты

    SpentOnWines: Сумма, потраченная на покупку вина за последние 2 года.
    SpentOnFruits: Сумма, потраченная на покупку фруктов за последние 2 года.
    SpentOnMeat: Сумма, потраченная на покупку мяса за последние 2 года.
    SpentOnFish: Сумма, потраченная на покупку рыбы за последние 2 года.
    SpentOnSweets: Сумма, потраченная на покупку сладостей за последние 2 года.
    SpentOnGold: Сумма, потраченная на покупку золота за последние 2 года.

##### Маркетинговые кампании

    DiscountedPurchases: Количество покупок со скидкой.
    AcceptedCampaign1: Указано, принял ли клиент предложение первой кампании (1 — да, 0 — нет).
    AcceptedCampaign2: Указано, принял ли клиент предложение второй кампании (1 — да, 0 — нет).
    AcceptedCampaign3: Указано, принял ли клиент предложение третьей кампании (1 — да, 0 — нет).
    AcceptedCampaign4: Указано, принял ли клиент предложение четвертой кампании (1 — да, 0 — нет).
    AcceptedCampaign5: Указано, принял ли клиент предложение пятой кампании (1 — да, 0 — нет).
    LastCampaignResponse: Указано, принял ли клиент предложение последней кампании (1 — да, 0 — нет).

##### Каналы покупок

    OnlinePurchases: Количество покупок, совершенных через веб-сайт компании.
    CatalogPurchases: Количество покупок, совершенных через каталог.
    InStorePurchases: Количество покупок, совершенных в магазине.
    WebsiteVisitsLastMonth: Количество визитов на веб-сайт компании за последний месяц.

##### Финансовые метрики

    ContactCost: Стоимость контакта с клиентом (постоянная метрика).
    RevenueGenerated: Доход, сгенерированный клиентом (постоянная метрика).

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics import silhouette_score, calinski_harabasz_score
from sklearn.compose import ColumnTransformer
from datetime import datetime
```

```python
def plot_distribution(column):
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    axes[0].hist(data[column], bins=range(data[column].min(), data[column].max() + 2),
                 edgecolor='black', align='left')
    axes[0].set_title(f'{column} Distribution', fontsize=14)
    axes[0].set_xlabel(column, fontsize=12)
    axes[0].set_ylabel('Count', fontsize=12)
    axes[0].grid(True, linestyle='--', alpha=0.7)
    axes[0].set_xticks(range(data[column].min(), data[column].max() + 1))


    data[column].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, counterclock=False, 
                         wedgeprops={'edgecolor': 'black'}, ax=axes[1])
    axes[1].set_title(column, fontsize=14)
    plt.show()
```

```python
def plot_num_dist(column):
    plt.hist(data[column], bins=50)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column, fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.xticks(rotation=0)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

```python
def plot_cat_dist(column):
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    data[column].value_counts().plot(kind='bar', ax=axes[0])
    axes[0].set_title(f'{column} Distribution', fontsize=14)
    axes[0].set_xlabel(column, fontsize=12)
    axes[0].set_ylabel('Count', fontsize=12)
    axes[0].grid(True, linestyle='--', alpha=0.7)
    
    data[column].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, counterclock=False, 
                         wedgeprops={'edgecolor': 'black'}, ax=axes[1])
    axes[1].set_title(column, fontsize=14)
    plt.show()
```

### Проведение кластеризации

```python
# Определение оптимального количества кластеров для KMeans с использованием метода локтя и силуэтного анализа
cluster_range = range(2, 11)
inertia = []
silhouette_scores = []
```

```python
for n_clusters in cluster_range:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init = 10)
    cluster_labels = kmeans.fit_predict(pca_data)
    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(pca_data, cluster_labels))
```

```python
fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.plot(cluster_range, inertia, marker='o', label='Inertia (Elbow Method)')
ax1.set_xlabel('Number of Clusters')
ax1.set_ylabel('Inertia', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

ax2 = ax1.twinx()
ax2.plot(cluster_range, silhouette_scores, marker='x', linestyle='--', color='green', label='Silhouette Score')
ax2.set_ylabel('Silhouette Score', color='green')
ax2.tick_params(axis='y', labelcolor='green')

fig.legend(loc="upper right", bbox_to_anchor=(0.9, 0.85))
plt.title('Elbow Method and Silhouette Analysis for Optimal Clusters')
plt.show()
```

Обе метрики указывают на то, что 2 кластера являются оптимальным выбором для данной задачи.

```python
kmeans = KMeans(n_clusters=2, random_state=42, n_init = 10)
cluster_labels = kmeans.fit_predict(pca_data)
silhouette_score(pca_data, cluster_labels)
```

```python
# Определение оптимального количества кластеров для AgglomerativeClustering с использованием силуэтного анализа
agglo_silhouette_scores = []

for k in cluster_range:
    agglo_clustering = AgglomerativeClustering(n_clusters=k, linkage='ward')
    agglo_labels = agglo_clustering.fit_predict(pca_data)
    score = silhouette_score(pca_data, agglo_labels)
    agglo_silhouette_scores.append(score)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(cluster_range, agglo_silhouette_scores, marker='o', color='purple')
plt.title('Silhouette Scores for Agglomerative Clustering')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.grid()
plt.show()
```

Для AgglomerativeClustering лучше всего подходит два кластера.

```python
agglo_clustering = AgglomerativeClustering(n_clusters=2, linkage='ward')
agglo_labels = agglo_clustering.fit_predict(pca_data)
silhouette_score(pca_data, agglo_labels)
```

```python
# Подберем коэффициенты для DBSCAN
eps_values = np.arange(0.5, 5.0, 0.5)
min_samples_values = range(3, 11)

# Store the silhouette scores for different combinations of eps and min_samples
dbscan_results = []
```

```python
for eps in eps_values:
    for min_samples in min_samples_values:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        labels = dbscan.fit_predict(pca_data)
        if len(set(labels) - {-1}) > 1:  
            silhouette = silhouette_score(pca_data[labels != -1], labels[labels != -1])
        else:
            silhouette = None
        dbscan_results.append((eps, min_samples, silhouette, len(set(labels) - {-1})))
```

```python
dbscan_results_df = pd.DataFrame(dbscan_results, columns=["eps", "min_samples", "silhouette_score", 'clusters'])
best_dbscan = dbscan_results_df.loc[dbscan_results_df['silhouette_score'].idxmax()]
```

### Формирование маркетинговых рекомендаций

Для Кластера 0 предположительно будут интересны  премиальные продукты и услуги, такие как отдых, путешествия, элитные рестораны и спа, подписки на премиальные сервисы (музыка, видео, обучение). Скорее всего этих клиентов интересуют инвестиционные и финансовые услуги, возможно консультации по вложениям, премиальные кредитные карты, страховые программы. Так же можно рекомендовать технологичные товары, например новые гаджеты, умные устройства для дома или электромобили. Возможно успешными будут эксклюзивные предложения, такие как закрытые мероприятия, клубы и лояльность для премиум-клиентов, ограниченные серии продуктов, например вин или ювелирных изделий. Так же можно предложить персонализированные скидки на премиальные сорта вина или подписки на винные дегустации, предложить комбинированные предложения, например, "Скидка на мясо и вино при совместной покупке", скидки на элитные продукты (органическое мясо, морепродукты, эксклюзивные сладости).

Стоит обратить внимание на ежедневные нужды и базовые услуги и предлагать программы лояльности для продуктовых магазинов и специальные предложения на товары первой необходимости. Рекомендуется создать программы поддержки семей со скидками на образовательные услуги (курсы, репетиторы), на семейные страховки.
Привлечь и удержать таких клиентов могут бонусы за активность, например акции «купи больше — плати меньше» и программы поощрения за регулярные покупки.

#### Как эти рекомендации помогут улучшить маркетинговую стратегию компании

Персонализация на основе покупательского поведения (например, интерес к премиальным товарам, инвестиционным услугам или эксклюзивным событиям) увеличит конверсию и отклик на маркетинговые кампании. Уменьшатся затраты на маркетинг за счет снижения “шумовых” рекламных воздействий, так как предложения будут релевантными для выбранной аудитории. Реализация реферальных программ, где существующие клиенты получают бонусы за привлечение новых клиентов, поможет расширить клиентскую базу. 
<br>Введение программ лояльности в продуктовых магазинах стимулирует постоянные покупки, обеспечивая регулярный доход компании. Программы «Купи больше — плати меньше» мотивируют клиентов совершать объемные покупки, увеличивая средний чек и общую выручку компании. Программы поощрения за регулярные покупки создают естественный механизм удержания клиентов, снижая расходы на повторный маркетинг. Акцент на семейные ценности и заботу о бюджете клиентов помогает выделиться среди конкурентов, которые могут не предлагать комплексных решений для семейного сегмента.
<br>Персонализированные предложения на основе анализа покупательских данных позволят предлагать клиентам только те товары и услуги, которые им наиболее интересны, тем самым увеличивая эффективность рекламы. Внедрение сезонных предложений (например, скидки перед школьным сезоном или новогодние акции) помогает поддерживать интерес клиентов на протяжении года.
<br>Внедрение предложенных рекомендаций позволит компании добиться следующих улучшений:
<br>1.	Рост лояльности и удовлетворенности клиентов.
<br>2.	Увеличение среднего чека и частоты покупок – благодаря выгодным программам и бонусам.
<br>3.	Снижение маркетинговых затрат – путем таргетированной персонализации и удержания постоянных клиентов.
<br>4.	Повышение конкурентоспособности – благодаря созданию репутации бренда.
<br>5.	Привлечение новых клиентов – через реферальные программы и сарафанное радио.
